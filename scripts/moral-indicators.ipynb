```
url = "https://moralfoundations.org/wp-content/uploads/files/downloads/moral%20foundations%20dictionary.dic"
moral_dict = pd.read_csv(url)
moral_dict.columns = ["raw"]
moral_dict["category"] = moral_dict.apply(lambda row: row["raw"].partition("\t")[2].replace("\t", ""), axis=1)
moral_dict["term"] = moral_dict.apply(lambda row: row["raw"].partition("\t")[0], axis=1)
moral_dict.tail(10)
```

```
### we are interested only in positive  terms in category 11 ("general")
negative_terms = ['wrong*',
 'evil',
 'immoral*',
 'bad',
 'offend*',
 'offensive*',
 'transgress*',
 'wicked*',
 'wretched*',
 ]
```

```
### var1 = only positive general 
def var1_selector(dataframe_row):
  if dataframe_row["category"] == "11":
    if dataframe_row["term"] not in negative_terms:
      value = True
    else: value = False
  else: value = False
  return value 

moral_dict["var1"] = moral_dict.apply(lambda row: var1_selector(row), axis=1)
moral_dict.tail(10) 
```

```
set_with_dataframe(sh.add_worksheet("moral-indicators", cols=1, rows=1), moral_dict)
```
